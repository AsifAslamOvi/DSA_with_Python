#  Exception Handling 

##  Definition

**Exception Handling** is a mechanism in Python used to handle runtime errors so that the program does not crash and can continue execution.

- In simple words:
It allows you to **handle errors gracefully** instead of stopping the program.

---

##  What is an Exception?

An **exception** is an error that occurs during the execution of a program.

###  Examples:

* Division by zero
* File not found
* Invalid input

---

#  Basic Syntax

```python
try:
    # Code that may cause error
except:
    # Code to handle error
```

---

##  Example

```python
try:
    x = int(input("Enter a number: "))
    print(10 / x)
except:
    print("An error occurred")
```

---

#  Multiple Exceptions

```python
try:
    a = int(input())
    b = int(input())
    print(a / b)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid input")
```

---

#  Using `else` Block

- Executes if no exception occurs

```python
try:
    x = 5
    y = 2
    result = x / y

except:
    print("Error")

else:
    print("Result:", result)
```

---

#  Using `finally` Block

- Always executes (whether error occurs or not)

```python
try:
    file = open("data.txt", "r")
except:
    print("File not found")
finally:
    print("Execution completed")
```

---

#  Full Structure

```python
try:
    # risky code

except ExceptionType:
    # handle specific error

else:
    # runs if no error

finally:
    # always runs
```

---

#  Catching Exception as Variable

```python
try:
    x = 10 / 0
except Exception as e:
    print("Error:", e)
```

---

#  Raising Exceptions

- You can create your own exceptions using `raise`

```python
def check_age(age):
    if age < 18:
        raise ValueError("Age must be 18 or above")
    return "Eligible"

print(check_age(16))
```

---

#  Common Built-in Exceptions

| Exception           | Description             |
| ------------------- | ----------------------- |
| `ZeroDivisionError` | Division by zero        |
| `ValueError`        | Invalid value           |
| `TypeError`         | Wrong data type         |
| `FileNotFoundError` | File not found          |
| `IndexError`        | List index out of range |

---

#  Advantages

1. Prevents program crash
2. Improves user experience
3. Helps debugging
4. Makes code more robust

---

#  Disadvantages

1. Overuse can hide real errors
2. Can make code complex
3. Performance overhead (slight)

---

#  Important Notes 

* Use `try` for risky code
* Use specific exceptions instead of general `except`
* `finally` always executes
* `else` runs only if no error occurs
* Avoid using bare `except:`

---

#  Real-Life Example

```python
try:
    file = open("data.txt", "r")
    content = file.read()
    print(content)

except FileNotFoundError:
    print("File does not exist")

finally:
    print("Done")
```

---

#  Summary

* Exception handling manages runtime errors
* Uses `try`, `except`, `else`, `finally`
* Prevents program crashes
* Improves reliability of programs

````python
## Exception try, except block
try:
    print("Enter a number: ")
    num = int(input())
    print("The number you entered is: ", num)
except ValueError:
    print("Invalid input. Please enter a valid number.")
````

````python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Please enter a non-zero denominator.")
except Exception as ex:
    print(ex)
````

````python
## try, except, else block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Please enter a non-zero denominator.")
else:
    print(f"The result is: {result}")
````

````python
## try, except, else , finally block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Please enter a non-zero denominator.")
except Exception as ex:
    print(ex)
else:
    print(f"The result is: {result}")
finally:
    print("This block will always be executed, regardless of whether an exception occurred or not.")
````

````python
## file handling and exception handling
try:
    file = open("non_existential_file.txt", "r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("The file you are trying to read does not exist.")
finally:
    if 'file' in locals() and not file.closed: 
        '''
        locals() -> returns a dictionary of the current local symbol table, we check if 'file' is in the local 
        symbol table and if it is not closed
        '''
        file.close()
        print("File closed successfully.")

````
