#  Linear Search 

##  Definition

**Linear Search** is a simple searching algorithm that checks each element in a list (or iterable) one by one until the desired element is found or the list ends.

- In simple words:
It **searches sequentially from start to end**.

---

##  How It Works

1. Start from the first element
2. Compare it with the target value
3. If matched → return index
4. If not → move to next element
5. Repeat until:

   * Element is found 
   * End of list reached 

---

##  Syntax (Conceptual)

```python
for each element in list:
    if element == target:
        return index
return -1
```

---

##  Basic Implementation

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

---

##  Example Usage

```python
numbers = [10, 20, 30, 40, 50]

result = linear_search(numbers, 30)

print("Index:", result)
```

### Output:

```
Index: 2
```

---

##  Step-by-Step Dry Run

Search for `30` in:

```
[10, 20, 30, 40]
```

| Step | Element | Comparison | Result   |
| ---- | ------- | ---------- | -------- |
| 1    | 10      | 10 ≠ 30    | Continue |
| 2    | 20      | 20 ≠ 30    | Continue |
| 3    | 30      | 30 = 30    | Found    |

- Index = 2

---

##  Using `for-each` Style

```python
def linear_search(arr, target):
    for index, value in enumerate(arr):
        if value == target:
            return index
    return -1
```

---

##  Searching in Strings

```python
text = "python"

for i, ch in enumerate(text):
    if ch == 'h':
        print("Found at index:", i)
```

---

##  Advantages

1. Simple and easy to implement
2. Works on unsorted data
3. No extra memory required
4. Suitable for small datasets

---

##  Disadvantages

1. Slow for large datasets
2. Checks every element in worst case
3. Not efficient compared to binary search

---

##  Time Complexity

| Case         | Complexity |
| ------------ | ---------- |
| Best Case    | O(1)       |
| Average Case | O(n)       |
| Worst Case   | O(n)       |

## Worst case occurs when element is:

* At the end
* Not present

---

##  Space Complexity

```
O(1)
```

- No extra memory used

---

##  Linear Search vs Binary Search

| Feature          | Linear Search | Binary Search    |
| ---------------- | ------------- | ---------------- |
| Data Requirement | Unsorted OK   | Must be sorted   |
| Speed            | Slow          | Fast             |
| Complexity       | O(n)          | O(log n)         |
| Implementation   | Easy          | Slightly complex |

---

##  Real-Life Example

```python
students = ["Asif", "Ovi", "Rahim", "Karim"]

name = "Rahim"

index = linear_search(students, name)

if index != -1:
    print("Found at index:", index)
else:
    print("Not found")
```

##  Multiple Occurrences

```python
def find_all(arr, target):
    indices = []
    for i in range(len(arr)):
        if arr[i] == target:
            indices.append(i)
    return indices
```

---

##  Important Notes 

* Works on both sorted and unsorted data
* Always checks elements one by one
* Inefficient for large datasets
* Useful when dataset is small or unsorted

---

#  Summary

* Linear search checks elements sequentially
* Returns index if found, otherwise -1
* Simple but slow for large data
* Time complexity: O(n)
* Best for small or unsorted datasets
