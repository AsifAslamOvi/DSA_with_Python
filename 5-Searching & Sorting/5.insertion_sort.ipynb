# Insertion Sort

##  Definition

**Insertion Sort** is a simple sorting algorithm that builds the final sorted list one element at a time by inserting each element into its correct position.

- In simple words:
It **takes one element and inserts it into the already sorted part of the list**.

---

##  How It Works

1. Assume the first element is already sorted
2. Take the next element
3. Compare it with elements in the sorted part
4. Shift larger elements to the right
5. Insert the element at the correct position
6. Repeat until all elements are sorted

---

##  Algorithm (Concept)

```python id="wvlq67"
for i from 1 to n-1:
    key = arr[i]
    j = i - 1
    
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j = j - 1
    
    arr[j + 1] = key
```

---

##  Basic Implementation

```python id="f5sm4b"
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        
        arr[j + 1] = key
    
    return arr
```

---

##  Advantages

1. Simple and easy to understand
2. Efficient for small datasets
3. Works well for nearly sorted data
4. Stable sorting algorithm
5. In-place (no extra memory needed)

---

##  Disadvantages

1. Slow for large datasets
2. Time complexity is high in worst case
3. Not suitable for big data

---

##  Time Complexity

| Case         | Complexity              |
| ------------ | ----------------------- |
| Best Case    | O(n) *(already sorted)* |
| Average Case | O(n²)                   |
| Worst Case   | O(n²)                   |

##  Insertion Sort vs Other Sorting

| Algorithm      | Best Case  | Worst Case | Notes                             |
| -------------- | ---------- | ---------- | --------------------------------- |
| Insertion Sort | O(n)       | O(n²)      | Good for small/nearly sorted data |
| Bubble Sort    | O(n)       | O(n²)      | More swaps                        |

#  Summary

* Insertion sort inserts elements into correct position
* Works efficiently for small or nearly sorted data
* Time complexity: O(n²), best case O(n)
* Space complexity: O(1)
* Stable and easy to implement


````python
def insertion_sort(arr):
    for current_pos in range(1, len(arr)):
        current_value = arr[current_pos]
        position = current_pos

        while position > 0 and arr[position - 1] > current_value:
            arr[position] = arr[position - 1]
            position -= 1
        arr[position] = current_value
    return arr


unsorted_list = [12, 25, 11, 34, 90, 27, 8, 19]
sorted_list = insertion_sort(unsorted_list)
print("Sorted list:", sorted_list)
````
