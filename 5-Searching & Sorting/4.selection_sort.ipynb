#  Selection Sort 

##  Definition

**Selection Sort** is a simple sorting algorithm that repeatedly finds the **minimum element** from the unsorted part of the list and places it at the beginning.

- In simple words:
It **selects the smallest element and puts it in the correct position**.

---

##  How It Works

1. Assume the first element is the minimum
2. Compare it with the rest of the elements
3. Find the actual minimum element
4. Swap it with the first element
5. Move to the next position and repeat
6. Continue until the list is sorted

---

##  Algorithm (Concept)

```python id="g3s8z2"
for i from 0 to n-1:
    min_index = i
    for j from i+1 to n:
        if arr[j] < arr[min_index]:
            min_index = j
    swap(arr[i], arr[min_index])
```

---

##  Basic Implementation

```python id="8v2xhf"
def selection_sort(arr):
    n = len(arr)
    
    for i in range(n):
        min_index = i
        
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        
        # Swap
        arr[i], arr[min_index] = arr[min_index], arr[i]
    
    return arr
```

##  Example Usage

```python id="ptoz3g"
numbers = [64, 25, 12, 22, 11]

sorted_numbers = selection_sort(numbers)
```

---

##  Advantages

1. Simple and easy to understand
2. Performs fewer swaps compared to bubble sort
3. No extra memory required (in-place)

---

##  Disadvantages

1. Slow for large datasets
2. Time complexity always O(n²)
3. Not stable (may change order of equal elements)

---

##  Time Complexity

| Case         | Complexity |
| ------------ | ---------- |
| Best Case    | O(n²)      |
| Average Case | O(n²)      |
| Worst Case   | O(n²)      |

- Same in all cases (no optimization benefit)

---

---

##  Selection Sort vs Bubble Sort

| Feature     | Selection Sort  | Bubble Sort    |
| ----------- | --------------- | -------------- |
| Swaps       | Fewer           | More           |
| Comparisons | Same            | Same           |
| Stability   | Not stable      | Stable         |
| Efficiency  | Slightly better | Slightly worse |

---

##  Real-Life Example

```python id="q4o3o0"
marks = [70, 50, 90, 60]

sorted_marks = selection_sort(marks)

print("Sorted marks:", sorted_marks)
```

---

##  Important Notes

* Always selects the minimum element
* Performs exactly (n-1) swaps
* Not adaptive (doesn't improve for sorted data)
* Easy to implement but inefficient for large data

---

#  Summary

* Selection sort finds minimum and places it correctly
* Works in-place with O(1) space
* Time complexity: O(n²)
* Performs fewer swaps than bubble sort
* Best for learning, not for large datasets


````python
def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr


unsorted_list = [12, 25, 11, 34, 90, 27, 8, 19]
sorted_list = selection_sort(unsorted_list)
print("Sorted list:", sorted_list)
````
