#  Bubble Sort

##  Definition

**Bubble Sort** is a simple sorting algorithm that repeatedly compares adjacent elements and swaps them if they are in the wrong order.

- In simple words:
It **pushes the largest element to the end in each pass**, just like bubbles rising to the surface.

---

##  How It Works

1. Compare the first two elements
2. Swap if they are in the wrong order
3. Move to the next pair
4. Repeat until the end of the list
5. Perform multiple passes until the list is sorted

---

##🔹 Algorithm (Concept)

```python id="mkh36p"
for i from 0 to n-1:
    for j from 0 to n-i-1:
        if arr[j] > arr[j+1]:
            swap(arr[j], arr[j+1])
```

---

##  Basic Implementation

```python id="0r6gk9"
def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
    return arr
```

##  Optimized Bubble Sort

- Stops early if no swaps occur (already sorted)

```python id="rg9m8k"
def bubble_sort_optimized(arr):
    n = len(arr)
    
    for i in range(n):
        swapped = False
        
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        
        if not swapped:
            break
    
    return arr
```

---

##  Advantages

1. Simple and easy to understand
2. No extra memory needed (in-place sorting)
3. Good for small datasets
4. Stable sorting algorithm

---

##  Disadvantages

1. Very slow for large datasets
2. Too many comparisons and swaps
3. Not efficient compared to other algorithms

---

##  Time Complexity

| Case         | Complexity                 |
| ------------ | -------------------------- |
| Best Case    | O(n) *(optimized version)* |
| Average Case | O(n²)                      |
| Worst Case   | O(n²)                      |

##  Bubble Sort vs Other Sorting

| Algorithm      | Time Complexity | Efficiency            |
| -------------- | --------------- | --------------------- |
| Bubble Sort    | O(n²)           | Slow                  |
| Selection Sort | O(n²)           | Slightly better       |
| Insertion Sort | O(n²)           | Better for small data |

#  Summary

* Bubble sort repeatedly swaps adjacent elements
* Works by pushing largest elements to the end
* Time complexity: O(n²)
* Space complexity: O(1)
* Best for small or educational use


````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

unsorted_list = [12, 25, 11, 34, 90, 27, 8, 19]
sorted_list = bubble_sort(unsorted_list)
print("Sorted list:", sorted_list)
````
