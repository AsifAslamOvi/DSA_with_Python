Operator

```python
# Python Operators: quick reference with examples

a, b = 10, 3

print("1) Arithmetic Operators")
print("a + b =", a + b)   # Addition
print("a - b =", a - b)   # Subtraction
print("a * b =", a * b)   # Multiplication
print("a / b =", a / b)   # Division (float)
print("a // b =", a // b) # Floor division
print("a % b =", a % b)   # Modulus (remainder)
print("a ** b =", a ** b) # Exponentiation

print("\n2) Comparison (Relational) Operators")
print("a == b:", a == b)  # Equal
print("a != b:", a != b)  # Not equal
print("a > b:", a > b)    # Greater than
print("a < b:", a < b)    # Less than
print("a >= b:", a >= b)  # Greater than or equal
print("a <= b:", a <= b)  # Less than or equal

print("\n3) Logical Operators")
x, y = True, False
print("x and y:", x and y) # True if both are True
print("x or y:", x or y)   # True if at least one is True
print("not x:", not x)     # Inverts boolean value

print("\n4) Assignment Operators")
n = 5
print("start n =", n)
n += 2
print("n += 2 ->", n)
n -= 1
print("n -= 1 ->", n)
n *= 3
print("n *= 3 ->", n)
n //= 2
print("n //= 2 ->", n)

print("\n5) Bitwise Operators")
p, q = 6, 3  # 6: 110, 3: 011
print("p & q =", p & q)   # AND
print("p | q =", p | q)   # OR
print("p ^ q =", p ^ q)   # XOR
print("~p =", ~p)         # NOT
print("p << 1 =", p << 1) # Left shift
print("p >> 1 =", p >> 1) # Right shift

print("\n6) Membership Operators")
nums = [1, 2, 3, 4]
print("2 in nums:", 2 in nums)
print("10 not in nums:", 10 not in nums)

print("\n7) Identity Operators")
list1 = [1, 2]
list2 = [1, 2]
list3 = list1
print("list1 is list2:", list1 is list2)       # Different objects
#list1 is list2 is False because both lists are stored in different memory locations.
#The "is" operator checks identity, not equality.
print("list1 is list3:", list1 is list3)       # Same object
print("list1 is not list2:", list1 is not list2)

print("\n8) Conditional (Ternary) Operator")
age = 20
status = "Adult" if age >= 18 else "Minor"
print("status =", status)
```
