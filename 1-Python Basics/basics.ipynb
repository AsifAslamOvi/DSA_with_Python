```python
# Basic Python syntax and semantics

# 1) Variables and basic data types
x = 10                  # int
pi = 3.14               # float
name = "Python"         # str
is_active = True        # bool
items = [1, 2, 3]       # list
point = (4, 5)          # tuple
unique = {1, 2, 3}      # set
person = {"age": 25}    # dict

print(x, pi, name, is_active)
print(items, point, unique, person)

# 2) Operators
a, b = 7, 3
print("Arithmetic:", a + b, a - b, a * b, a / b, a // b, a % b, a ** b)
print("Comparison:", a > b, a == b, a != b)
print("Logical:", (a > 5 and b < 5), (a < 5 or b < 5), not (a == b))

# 3) Conditional statements
score = 82
if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
else:
    grade = "C"
print("Grade:", grade)

# 4) Loops
for i in range(3):
    print("for loop:", i)

n = 0
while n < 3:
    print("while loop:", n)
    n += 1

# 5) Functions
def add(p, q=0):
    return p + q

print("Function:", add(5, 2), add(5))

# 6) Scope and semantics (mutable vs immutable)
num = 10          # immutable
lst = [1, 2]      # mutable

def modify(n, l):
    n += 1        # does not change outer num
    l.append(3)   # changes outer list
    return n, l

new_num, new_lst = modify(num, lst)
print("num:", num, "new_num:", new_num)
print("lst:", lst, "new_lst:", new_lst)

# 7) Exception handling
try:
    value = int("123")
    print("Parsed:", value)
except ValueError:
    print("Invalid number")

# 8) Class (object-oriented basics)
class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} says woof"

d = Dog("Buddy")
print(d.speak())
```

```python
indentation_notes = """
Python uses indentation (leading whitespace) to define code blocks.

Key points:
1. Use 4 spaces per indentation level (PEP 8).
2. Blocks after statements like if, for, while, def, and class must be indented.
3. Consistent indentation is required, or Python raises IndentationError.
4. Avoid mixing tabs and spaces.

Example:
if True:
    print("inside block")
print("outside block")
"""

print(indentation_notes)
```

```python
line_continuation_notes = """
Line continuation in Python lets one logical statement span multiple lines.

Two common ways:

1) Explicit line continuation with backslash (\)
    - Put \\ at the end of the line.
    - Nothing (not even spaces/comments) should follow the backslash.

2) Implicit line continuation inside (), [], {}
    - Preferred style (cleaner and safer).
    - No backslash needed.

Examples:
"""

# 1) Explicit continuation
total_explicit = 10 + 20 + 30 + \
                      40 + 50

# 2) Implicit continuation (preferred)
total_implicit = (
     10 + 20 + 30 +
     40 + 50
)

nums = [
     1, 2, 3,
     4, 5, 6
]

print(line_continuation_notes)
print("Explicit total:", total_explicit)
print("Implicit total:", total_implicit)
print("List with implicit continuation:", nums)
```

```python
type_inference_notes = """
Type inference is when Python determines a value's type automatically from the assigned expression.

In Python:
- You usually do not declare variable types explicitly.
- The interpreter infers types at runtime.

Examples:
"""

# Python infers these types from assigned values
count = 42            # inferred as int
price = 19.99         # inferred as float
title = "Notebook"    # inferred as str
flags = [True, False] # inferred as list

print(type_inference_notes)
print("count ->", count, "| type:", type(count).__name__)
print("price ->", price, "| type:", type(price).__name__)
print("title ->", title, "| type:", type(title).__name__)
print("flags ->", flags, "| type:", type(flags).__name__)

# Reusing existing notebook variables
print("nums ->", nums, "| type:", type(nums).__name__)
print("total_explicit and total_implicit have same value:",
    total_explicit == total_implicit)
```
