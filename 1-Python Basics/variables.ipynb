```python
import copy

"""
Python Variables: quick reference with examples.

A variable is a name bound to an object in memory.
Python is dynamically typed: variable types are decided at runtime.
"""

# 1) Basic assignment
name = "Alice"          # str
age = 30                # int
height = 1.68           # float
is_active = True        # bool

# 2) Multiple assignment
x, y, z = 1, 2, 3

# 3) Same value to multiple variables
a = b = c = 0

# 4) Naming rules
# valid: letters, digits, underscore; cannot start with digit
user_name = "bob"
_user_id = 101
# 2user = "invalid"      # SyntaxError

# 5) Case-sensitive
value = 10
Value = 20  # different variable

# 6) Reassignment (type can change)
var = 100
var = "now I am a string"

# 7) Check type
print(type(name), type(age), type(height), type(is_active))

# 8) Mutable vs immutable
# Immutable example (int, str, tuple): new object on change
n = 5
print("n id before:", id(n))
n = n + 1
print("n id after :", id(n))

# Mutable example (list, dict, set): can change in place
items = [1, 2, 3]
print("items id before:", id(items))
items.append(4)
print("items id after :", id(items), items)

# 9) Object references (important behavior)
list1 = [10, 20]
list2 = list1          # both reference same list
list2.append(30)
print("list1:", list1) # [10, 20, 30]

# 10) Copy to avoid shared reference
original = [[1, 2], [3, 4]]
shallow = original.copy()          # shallow copy
deep = copy.deepcopy(original)     # deep copy

original[0].append(99)
print("original:", original)  # [[1, 2, 99], [3, 4]]
print("shallow :", shallow)   # inner list changed too
print("deep    :", deep)      # unaffected

# 11) Constants by convention
PI = 3.14159  # Python has no true constant keyword

# 12) Local / global scope
global_var = "I am global"

def demo_scope():
    local_var = "I am local"
    print(global_var)
    print(local_var)

demo_scope()

# 13) Deleting variable binding
temp = "to be removed"
del temp
# print(temp)  # NameError

print("Done.")
```

```python
import ast

# 1) String input
inp_str = input("Enter a string: ")

# 2) Integer input
inp_int = int(input("Enter an integer: "))

# 3) Float input
inp_float = float(input("Enter a float: "))

# 4) Boolean input (custom parsing)
inp_bool_text = input("Enter boolean (true/false): ").strip().lower()
inp_bool = inp_bool_text in ("true", "1", "yes", "y")

# 5) Multiple values in one line (as strings)
s1, s2 = input("Enter two words (space-separated): ").split()

# 6) Multiple integers in one line
i1, i2 = map(int, input("Enter two integers (space-separated): ").split())

# 7) List of integers
inp_list_int = list(map(int, input("Enter list of integers (space-separated): ").split()))

# 8) Tuple of integers
inp_tuple_int = tuple(map(int, input("Enter tuple values (space-separated integers): ").split()))

# 9) Set of integers
inp_set_int = set(map(int, input("Enter set values (space-separated integers): ").split()))

# 10) List of characters from one string
inp_chars = list(input("Enter a word to split into characters: "))

# 11) Dict / list / nested structures using literal_eval
def safe_literal_eval(prompt, default_value, expected_type):
	raw = input(prompt).strip()
	try:
		parsed = ast.literal_eval(raw)
		if not isinstance(parsed, expected_type):
			print(f"Warning: expected {expected_type.__name__}, got {type(parsed).__name__}. Using default.")
			return default_value
		return parsed
	except (SyntaxError, ValueError):
		print(f"Warning: invalid literal syntax. Using default {default_value}.")
		return default_value

inp_dict = safe_literal_eval("Enter a dictionary (e.g. {'a': 1, 'b': 2}): ", {}, dict)
inp_nested_list = safe_literal_eval("Enter nested list (e.g. [[1,2],[3,4]]): ", [], list)

# 12) Multiple lines input
n_lines = int(input("How many lines to read? "))
inp_lines = [input(f"Line {idx + 1}: ") for idx in range(n_lines)]

print("\n--- Parsed values and types ---")
print("inp_str:", inp_str, type(inp_str))
print("inp_int:", inp_int, type(inp_int))
print("inp_float:", inp_float, type(inp_float))
print("inp_bool:", inp_bool, type(inp_bool))
print("s1, s2:", s1, s2, type(s1), type(s2))
print("i1, i2:", i1, i2, type(i1), type(i2))
print("inp_list_int:", inp_list_int, type(inp_list_int))
print("inp_tuple_int:", inp_tuple_int, type(inp_tuple_int))
print("inp_set_int:", inp_set_int, type(inp_set_int))
print("inp_chars:", inp_chars, type(inp_chars))
print("inp_dict:", inp_dict, type(inp_dict))
print("inp_nested_list:", inp_nested_list, type(inp_nested_list))
print("inp_lines:", inp_lines, type(inp_lines))
```
