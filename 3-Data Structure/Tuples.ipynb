Tuples in Python (Theory)
# Definition
    A tuple is an ordered, immutable collection of elements in Python.
    Ordered → Elements have a fixed position (indexing works)
    Immutable → Cannot be changed after creation
    Allows duplicate values
# Syntax
    Tuples are created using parentheses ():
    t = (1, 2, 3)
## Single element tuple (important!)
    t = (5,)   # comma is mandatory
## Characteristics of Tuples
    1) Ordered
    2) Elements maintain insertion order
    3) Immutable
    4) Cannot modify, add, or delete elements
    5) Allow Duplicates
        (1, 2, 2, 3)
    6) Heterogeneous
    7) Can store different data types
        (1, "Hello", 3.5, True)
## Accessing Elements
    ✔ Indexing
        t = (10, 20, 30)
        print(t[1])  # 20
    ✔ Negative Indexing
        print(t[-1])  # 30
    ✔ Slicing
        print(t[0:2])  # (10, 20)
## Tuple Operations
    ✔ Concatenation
        t1 = (1, 2)
        t2 = (3, 4)
        print(t1 + t2)  # (1, 2, 3, 4)
    ✔ Repetition
        print((1, 2) * 2)  # (1, 2, 1, 2)
## Tuple Methods
    Tuples have very few methods (because they are immutable):
        => count()
            t = (1, 2, 2, 3)
            t.count(2)  # 2
        => index()
            t.index(3)  # 3rd position
## Packing and Unpacking
    ✔ Packing
        t = 1, 2, 3
    ✔ Unpacking
        a, b, c = (1, 2, 3)
## Immutability of Tuples
    t = (1, 2, 3)
    t[0] = 10  # ❌ Error
    But:
        t = (1, [2, 3])
        t[1][0] = 100  # Allowed (list inside tuple)

# Tuple itself is immutable, but mutable elements inside it can change

## Advantages of Tuples
    ✔ Faster than lists
    ✔ Memory efficient
    ✔ Data safety (cannot be changed)
    ✔ Can be used as dictionary keys

```python
## creating a tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
numbers = tuple([1, 2, 3, 4, 5, 6])
print(numbers)
```

```python
list((1, 2, 3, 4, 5, 6))
```

```python
mixed_tuple = (1, "hello", 3.14, [1, 2, 3], (4, 5), True)
print(mixed_tuple)
```

```python
## Accessing tuple elements
numbers = (1, 2, 3, 4, 5, 6, 7)
print(numbers[-1])
print(numbers[0:3])
print(numbers[::2])
print(numbers[::-1])
```

```python
## Tuples operations

# Concatenation
print(numbers + mixed_tuple)

# Repetition
print(numbers * 2)
print(mixed_tuple * 3)
```

```python
## Immutable nature of tuples

lst = [1, 2, 3, 4, 5]
print(lst)

lst[1] = "Ovi"
print(lst)

# but for tuples, we cannot change the elements
# numbers[1] = "Ovi" # TypeError: 'tuple' object does not support item assignment
# print(numbers)
try:
    numbers[1] = "Ovi"
except TypeError as e:
    print(e)

```

```python
## Tuples Methods
print(numbers.count(3))
print(numbers.index(4)) # 3 is the index of the first occurrence of 4 in the tuple
```

```python
## packing and unpacking tuples

# packing
person = ("Ovi", 22, "Engineer")
print(person)

# unpacking
name, age, profession = person
print(name)
print(age)
print(profession)
```

```python
## unpacking with * operator
numbers = (1, 2, 3, 4, 5, 6, 7)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested lists
lst = [[1, 2, 3], [4, 5, 6], [1, "Ovi", 3.80, "Engineer"]]
print(lst[0][0:2])
print(lst[2][1])

## Netsed tuples
nested_tuple = ((1, 2, 3), (4, 5, 6), (1, "Ovi", 3.80, "Engineer"))
print(nested_tuple[0][0:2])
print(nested_tuple[2][1])

## Iterating through tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item)
    print("-----")
```
