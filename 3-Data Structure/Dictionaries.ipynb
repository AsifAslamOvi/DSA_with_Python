# Python Dictionaries

A **dictionary** is a built-in Python data structure that stores data as **key–value pairs**.

```python
person = {"name": "Ada", "age": 28, "is_engineer": True}
```

- Keys are unique.
- Keys are typically immutable types (`str`, `int`, `tuple` of immutables, etc.).
- Values can be any Python object.
- Dictionaries are mutable (you can add, update, remove items).
- In modern Python, dictionaries preserve insertion order.

---

## Creating Dictionaries

```python
# Literal
d1 = {"a": 1, "b": 2}

# dict constructor
d2 = dict(a=1, b=2)

# From list of tuples
d3 = dict([("a", 1), ("b", 2)])

# Empty dictionary
d4 = {}
```

---

## Accessing Values

```python
d = {"name": "Ada", "age": 28}

d["name"]          # "Ada" (KeyError if missing)
d.get("age")       # 28
d.get("email")     # None
d.get("email", "") # default value if missing
```

---

## Adding and Updating

```python
d = {"a": 1}
d["b"] = 2              # add new key
d["a"] = 10             # update existing key
d.update({"c": 3})      # bulk update
d.update(d=4, e=5)      # keyword-style update
```

---

## Removing Items

```python
d = {"a": 1, "b": 2, "c": 3}

del d["a"]          # remove by key (KeyError if missing)
value = d.pop("b")  # remove + return value
item = d.popitem()  # remove + return last inserted pair
d.clear()           # remove all items
```

---

## Common Dictionary Methods

```python
d.keys()      # view of keys
d.values()    # view of values
d.items()     # view of (key, value) pairs
d.copy()      # shallow copy
d.setdefault("x", 0)  # get existing or set default
```

---

## Iteration

```python
d = {"a": 1, "b": 2, "c": 3}

for k in d:
    print(k)              # keys

for v in d.values():
    print(v)              # values

for k, v in d.items():
    print(k, v)           # key-value pairs
```

---

## Membership Checks

```python
"a" in d         # checks keys
1 in d.values()  # checks values
```

---

## Dictionary Comprehensions

```python
squares = {x: x*x for x in range(5)}
# {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

With condition:

```python
evens = {x: x*x for x in range(10) if x % 2 == 0}
```

---

## Nested Dictionaries

```python
users = {
    "u1": {"name": "Ada", "age": 28},
    "u2": {"name": "Linus", "age": 35}
}

users["u1"]["name"]  # "Ada"
```

---

## Copying: Shallow vs Deep

```python
import copy

original = {"a": [1, 2]}
shallow = original.copy()
deep = copy.deepcopy(original)
```

- `copy()` duplicates only the top-level dictionary.
- Nested objects are still shared in a shallow copy.
- Use `deepcopy()` for full independent copies.

---

## Merging Dictionaries

```python
a = {"x": 1, "y": 2}
b = {"y": 99, "z": 3}

merged1 = {**a, **b}   # b overrides duplicate keys
merged2 = a | b        # Python 3.9+
a |= b                 # in-place merge
```

---

## Dictionary Views

`keys()`, `values()`, and `items()` return dynamic view objects:

```python
d = {"a": 1}
kv = d.keys()
print(kv)      # dict_keys(['a'])
d["b"] = 2
print(kv)      # dict_keys(['a', 'b'])  <- updated automatically
```

---

## Performance Notes

- Average-case lookup, insert, delete: **O(1)**.
- Dictionaries are implemented as hash tables.
- Good hashability and key choice matter for performance.

---

## Typical Use Cases

- Fast lookups by identifier.
- Counting frequencies.
- Grouping records.
- JSON-like data handling.
- Caching and memoization.
- Configuration objects.

Example: frequency counter

```python
text = "banana"
freq = {}
for ch in text:
    freq[ch] = freq.get(ch, 0) + 1
# {'b': 1, 'a': 3, 'n': 2}
```

---

## Best Practices

- Prefer `get()` when keys may be missing.
- Use `items()` when both key and value are needed.
- Use clear, consistent key names.
- Avoid mutating a dictionary while iterating over it directly.
- Use `defaultdict` (from `collections`) for default-value patterns.

```python
from collections import defaultdict

groups = defaultdict(list)
for name, dept in [("Ada", "R&D"), ("Bob", "HR"), ("Eve", "R&D")]:
    groups[dept].append(name)
```

---

## Related Types

- `collections.defaultdict`: automatic default values.
- `collections.Counter`: counting hashable objects.
- `collections.OrderedDict`: mostly legacy since standard dict preserves order.
- `types.MappingProxyType`: read-only dictionary view.

## Summary

Python dictionaries are flexible, efficient mappings from unique keys to values. They are central to everyday Python programming for structured data, quick lookups, and clean data transformations.

````python
## creating dictionaries 
empty_dict = {}
print(type(empty_dict)) # <class 'dict'>
````

````python
empty_dict2 = dict()
print(type(empty_dict2)) # <class 'dict'>
````

````python
student = {
    'name': 'John Doe',
    'age': 20,
    'courses': ['Math', 'CompSci'] 
}

print(student)
print(type(student)) # <class 'dict'>
````

````python
student = {
    'name': 'John Doe',
    'age': 20,
    'name': ['Math', 'CompSci'] ## duplicate key, the last one will be used
}

print(student)
````

````python
## accessing dictionary elements
student = {
    'name': 'John Doe',
    'age': 20,
    'courses': ['Math', 'CompSci'] 
}

print(student)

print(student['age'])

## Accessing dictionary elements using get() method
print(student.get('age')) # 20
print(student.get('phone')) # None
print(student.get('phone', 'Not Found')) # Not Found
````

````python
## Modifying dictionary elements
student['age'] = 21
print(student)
student['phone'] = '01771158082'
print(student)
del student['courses'][1]
print(student)
````

````python
## Dictionary methods

keys = student.keys()
print(keys)
Values = student.values()
print(Values)
items = student.items()
print(items)
````

````python
## shallow copy of a dictionary
student2 = student.copy()
print(student2)
student2['name'] = 'Jane Doe'
print(student2)
print(student)
````

````python
## iterating over keys
print("Keys in student dictionary:")
for key in student.keys(): 
    print(key)

## iterating over values
print("\nValues in student dictionary:")
for value in student.values():
    print(value)

## iterating over key-value pairs
print("\nKey-value pairs in student dictionary:")
for key, value in student.items():
    print(f"{key} : {value}")

````

````python
## Nested dictionaries
students = {
    'student1': {
        'name': 'John Doe',
        'age': 20,
        'courses': ['Math', 'CompSci']
    },
    'student2': {
        'name': 'Jane Doe',
        'age': 22,
        'courses': ['History', 'Art']
    }
}
print(students)
````

````python
## Access nested dictionary elements
print(students['student1']['name']) # John Doe
print(students['student2']['courses'][0]) # History
````

````python
## Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"Student ID: {student_id}")
    for key, value in student_info.items():
        print(f"{key} : {value}")
    print()
````

````python
## Dictionary comprehension
squares = {x: x**2 for x in range(1, 6)}
print(squares)
````

````python
## conditional dictionary comprehension
even_squares = {x: x**2 for x in range(1, 11) if x%2 == 0}
print(even_squares)
````

````python
## practical example of dictionary comprehension
numbers = [1, 2, 2, 3, 3, 3, 4, 5, 5, 6]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
````

````python
## Merge 2 dictionaries
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
````
