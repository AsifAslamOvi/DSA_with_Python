Sets

```python
# Python Sets - quick guide with examples

# 1) Create sets
a = {1, 2, 3, 3, 2}          # duplicates are removed automatically
b = set([3, 4, 5])           # from list
empty_set = set()            # {} creates an empty dict, not a set

print("a:", a)               # {1, 2, 3}
print("b:", b)               # {3, 4, 5}
print("empty_set:", empty_set)

# 2) Basic properties
# - Unordered: no index access (a[0] is invalid)
# - Mutable: can add/remove items
# - Elements must be hashable (immutable types like int, str, tuple)

# 3) Membership test (fast lookup)
print("2 in a:", 2 in a)
print("10 in a:", 10 in a)

# 4) Add / update
a.add(10)                    # add one element
a.update([20, 30], {40, 50}) # add multiple elements from iterables
print("after add/update a:", a)

# 5) Remove elements
a.remove(10)                 # raises KeyError if not present
a.discard(999)               # no error if missing
popped = a.pop()             # removes arbitrary element
print("popped:", popped)
print("after remove/discard/pop a:", a)

# 6) Set operations
x = {1, 2, 3, 4}
y = {3, 4, 5, 6}

print("union:", x | y)                    # {1,2,3,4,5,6}
print("intersection:", x & y)             # {3,4}
print("difference x-y:", x - y)           # {1,2}
print("difference y-x:", y - x)           # {5,6}
print("symmetric difference:", x ^ y)     # {1,2,5,6}

# method style (same results)
print("x.union(y):", x.union(y))
print("x.intersection(y):", x.intersection(y))

# 7) Subset / superset / disjoint
m = {1, 2}
n = {1, 2, 3}
z = {7, 8}

print("m subset of n:", m.issubset(n))
print("n superset of m:", n.issuperset(m))
print("m disjoint with z:", m.isdisjoint(z))

# 8) Set comprehension
squares = {i * i for i in range(6)}
print("set comprehension (squares):", squares)

# 9) frozenset (immutable set)
fs = frozenset([1, 2, 3])
print("frozenset:", fs)
# fs.add(4)  # AttributeError (immutable)

# 10) Practical examples

# Remove duplicates from list
nums = [1, 2, 2, 3, 4, 4, 5]
unique_nums = list(set(nums))
print("unique nums:", unique_nums)

# Common items between two lists
list1 = ["python", "java", "go"]
list2 = ["rust", "python", "go"]
common = set(list1) & set(list2)
print("common items:", common)

# Difference between two collections
only_in_list1 = set(list1) - set(list2)
print("only in list1:", only_in_list1)
```

```python
## Mathematical Operation 
Set1 = {1, 2, 3, 4, 5}
Set2 = {4, 5, 6, 7, 8}

## union
union_set = Set1.union(Set2)
print("Union:", union_set)

## Intersection
intersection_set = Set1.intersection(Set2)
print("Intersection:", intersection_set)

## Difference
difference_set = Set1.difference(Set2)
print("Difference (Set1 - Set2):", difference_set)

## intersection update -> modifies Set1 to keep only elements also in Set2
Set1.intersection_update(Set2)
print("Set1 after intersection update:", Set1)

print("Set1 after intersection update:", Set1)
print("Set2 remains unchanged:", Set2)
```

```python
## Symmetric Difference
sym_diff_set = Set1.symmetric_difference(Set2)
print("Symmetric Difference:", sym_diff_set)
```

```python
## Sets Methods
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5}

## is subset
print("set1 is subset of set2:", set1.issubset(set2))
## is superset
print("set1 is superset of set2:", set1.issuperset(set2))
```

```python
## counting unique words in a text
text = "In this section we are talking about sets in python. Sets are a collection of unique elements."
words = text.split()

## converts the list of words to a set to get unique words
unique_words = set(words)
print("Unique words in the text:", unique_words)
print("Number of unique words:", len(unique_words))
```
