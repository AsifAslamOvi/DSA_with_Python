Lists

```python
# Python Lists: quick, practical overview with examples

# 1) Create lists
numbers = [10, 20, 30, 40]
mixed = [1, "hello", 3.14, True]
empty = []

print("numbers:", numbers)
print("mixed:", mixed)
print("empty:", empty)

# 2) Access elements (indexing)
print("\nFirst element:", numbers[0])   # 10
print("Last element:", numbers[-1])     # 40

# 3) Slicing
print("\nSlice [1:3]:", numbers[1:3])   # [20, 30]
print("Slice [:2]:", numbers[:2])       # [10, 20]
print("Slice [::2]:", numbers[::2])     # [10, 30]

# 4) Modify lists (lists are mutable)
numbers[1] = 25
print("\nAfter changing index 1:", numbers)

# 5) Common list methods
numbers.append(50)              # add at end
numbers.extend([60, 70])        # add multiple
numbers.insert(1, 15)           # insert at position
print("\nAfter append/extend/insert:", numbers)

numbers.remove(25)              # remove first matching value
popped = numbers.pop()          # remove and return last item
print("After remove/pop:", numbers)
print("Popped value:", popped)

# 6) Useful operations
print("\nLength:", len(numbers))
print("Membership (30 in numbers):", 30 in numbers)
print("Count of 10:", numbers.count(10))
print("Index of 30:", numbers.index(30))

# 7) Sorting and reversing
data = [5, 2, 9, 1]
print("\nOriginal data:", data)
print("sorted(data):", sorted(data))  # returns new sorted list
data.sort()                           # sorts in place
print("After data.sort():", data)
data.reverse()
print("After data.reverse():", data)

# 8) Copying lists
a = [1, 2, 3]
b = a          # same reference
c = a.copy()   # shallow copy
a.append(4)
print("\nOriginal a:", a)
print("b (same object as a):", b)
print("c (copied before append):", c)

# 9) Looping through a list
print("\nLooping with index and value:")
for i, value in enumerate(numbers):
    print(i, value)

# 10) List comprehensions
squares = [x * x for x in range(6)]
evens = [x for x in range(10) if x % 2 == 0]
print("\nSquares:", squares)
print("Even numbers:", evens)

# 11) Nested lists
matrix = [[1, 2], [3, 4], [5, 6]]
print("\nNested list (matrix):", matrix)
print("matrix[1][0]:", matrix[1][0])  # 3
```

```python
### list comprehension with function calls
words = ["hello", "world", "python" , "list", "comprehension"]
lengths = [len(word) for word in words]
print("\nWords:", words, "Lengths:", lengths)
```
