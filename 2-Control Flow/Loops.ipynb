Loops

```python
# Python loops: practical overview with examples

# 1) for loop: iterate over a sequence
print("1) for loop over a list")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# 2) range(): common numeric loop
print("\n2) for loop with range()")
for i in range(1, 6):  # 1 to 5
    print(i, end=" ")
print()

# 3) while loop: runs while condition is True
print("\n3) while loop")
count = 1
while count <= 3:
    print("count =", count)
    count += 1

# 4) break: stop loop early
print("\n4) break example")
for n in range(10):
    if n == 5:
        break
    print(n, end=" ")
print()

# 5) continue: skip current iteration
print("\n5) continue example")
for n in range(6):
    if n % 2 == 0:
        continue
    print(n, end=" ")
print()

# 6) else with loops: runs if loop finishes normally (no break)
print("\n6) loop else example")
for n in range(3):
    print("checking", n)
else:
    print("loop completed without break")

# 7) nested loops
print("\n7) nested loops (multiplication table 1-3)")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i}x{j}={i*j}", end="  ")
    print()

# 8) list comprehension: compact loop syntax
print("\n8) list comprehension")
squares = [x * x for x in range(6)]
print("squares:", squares)

# 9) dictionary iteration
print("\n9) loop through dictionary")
student = {"name": "Asha", "age": 20, "grade": "A"}
for key, value in student.items():
    print(f"{key}: {value}")
```
