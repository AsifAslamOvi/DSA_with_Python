#  Python Standard Library

##  Definition

The **Python Standard Library** is a collection of built-in modules that come pre-installed with Python. These modules provide ready-to-use functions and tools for common programming tasks.

- In simple words:
It is a **large set of pre-written code that saves time and effort**.

---

##  Why Standard Library is Important?

* No need to install extra packages
* Saves development time
* Reliable and optimized code
* Used in almost every Python project
* Covers many domains (math, file handling, OS, networking, etc.)

---

#  Common Standard Library Modules

##  1. `math` Module

Provides mathematical functions.

```python id="m4h1xq"
import math

print(math.sqrt(25))
print(math.pow(2, 3))
print(math.factorial(5))
```

---

##  2. `random` Module

Used to generate random numbers.

```python id="v3q8ab"
import random

print(random.randint(1, 10))
print(random.random())
```

---

##  3. `datetime` Module

Used for working with dates and time.

```python id="k9f3ds"
import datetime

print(datetime.datetime.now())
```

---

##  4. `os` Module

Used for interacting with the operating system.

```python id="u2p9gh"
import os

print(os.getcwd())  # current directory
```

---

##  5. `sys` Module

Provides system-specific parameters and functions.

```python id="x7m2pl"
import sys

print(sys.version)
print(sys.path)
```

---

##  6. `time` Module

Used for time-related tasks.

```python id="c1r8nv"
import time

print(time.time())  # current timestamp
```

---

##  7. `re` Module (Regular Expressions)

Used for pattern matching.

```python id="r9t2qa"
import re

pattern = r"\d+"
text = "I have 2 apples and 10 oranges"

print(re.findall(pattern, text))
```

---

##  8. `collections` Module

Provides specialized data structures.

```python id="l3w8zx"
from collections import Counter

data = [1, 2, 2, 3, 3, 3]
print(Counter(data))
```

---

##  9. `json` Module

Used for working with JSON data.

```python id="n8q1ty"
import json

data = {"name": "Asif", "age": 21}

json_str = json.dumps(data)
print(json_str)
```

---

##  10. `itertools` Module

Used for advanced iteration tools.

```python id="p5s9wd"
import itertools

nums = [1, 2, 3]
print(list(itertools.permutations(nums)))
```

#  Standard Library vs External Libraries

| Feature      | Standard Library  | External Libraries |
| ------------ | ----------------- | ------------------ |
| Installation | Not needed        | Required (pip)     |
| Source       | Comes with Python | Third-party        |
| Examples     | math, os, sys     | numpy, pandas      |
| Reliability  | Very high         | Varies             |

````python
## High level operations on files and directories
import shutil
shutil.copy('source.txt', 'destination.txt') # copy a file from source to destination
````

````python
## Data Serialization
import json
data = {'name': 'Ovi', 'age': 23, 'city': 'Zurich'}

json_string = json.dumps(data) # Convert Python object to JSON string
print(json_string)
print(type(json_string))

loaded_data = json.loads(json_string) # Convert JSON string back to Python object
print(loaded_data)
print(type(loaded_data))
````

````python
## csv module for reading and writing CSV files
import csv
with open('example.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['Ovi', 23, 'Zurich'])

with open('example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
````

````python
## datetime module for working with dates and times
from datetime import datetime, timedelta

now = datetime.now()
print("Current date and time:", now)

future_date = now + timedelta(days=7)
print("Date and time after 7 days:", future_date)
````

````python
### time module for working with time-related functions
import time
print("Current time: ", time.ctime())
print(time.time()) # Get current time in seconds since the epoch

time.sleep(2) # Sleep for 2 seconds
print("This message is printed after a 2-second delay.")
print(time.time()) # Get current time again to see the delay
````

````python
## Regular Expressions
import re

pattern = r'\d+' # d -> digit, + -> one or more occurrences
text = "There are 3 cats and 4 dogs."

matches = re.search(pattern, text)
if matches:
    print("First number found:", matches.group())
matches = re.findall(pattern, text)
print("All numbers found:", matches)
````
