#  Importing Modules and Packages 

##  Definition

###  Module

A **module** is a file containing Python code (functions, variables, classes) that can be reused in other programs.

- Example: `math.py`, `random.py`

---

###  Package

A **package** is a collection of related modules organized in directories.

- It usually contains an `__init__.py` file.

---

##  Why Use Modules and Packages?

* Code reusability
* Better organization
* Avoid repetition
* Easier maintenance
* Modular programming

---

#  Importing Modules

##  1. Import Entire Module

```python
import math

print(math.sqrt(16))
```

- Access using `module_name.function_name`

---

##  2. Import Specific Functions

```python
from math import sqrt, pow

print(sqrt(25))
print(pow(2, 3))
```

- No need to use module name

---

##  3. Import with Alias

```python
import math as m

print(m.sqrt(36))
```

- Shortens module name

---

##  4. Import All (*Not Recommended*)

```python
from math import *

print(sqrt(49))
```

❌ Can cause confusion and name conflicts

---

#  Importing Custom Modules

##  Example: Create a Module

### 📁 `my_module.py`

```python
def greet(name):
    return f"Hello, {name}"

def add(a, b):
    return a + b
```

---

##  Use the Module

```python
import my_module

print(my_module.greet("Asif"))
print(my_module.add(3, 4))
```

---

#  Packages in Python

##  Structure

```
my_package/
│
├── __init__.py
├── module1.py
└── module2.py
```

---

##  Import from Package

```python
from my_package import module1

module1.function_name()
```

---

##  Import Specific Function from Package

```python
from my_package.module1 import function_name
```

---

#  Relative vs Absolute Import

##  Absolute Import

```python
from my_package.module1 import function_name
```

---

##  Relative Import

```python
from . import module1
from ..subpackage import module2
```

- Used inside packages

#  Module Search Path

Python searches for modules in:

1. Current directory
2. Installed libraries
3. System paths

```python
import sys
print(sys.path)
```

---

#  Important Notes 

* Modules = single file
* Packages = collection of modules
* Use `import module` for full access
* Use `from module import name` for specific items
* Avoid `import *`
* Use alias (`as`) for readability

---

#  Advantages

1. Code reusability
2. Better organization
3. Easier debugging
4. Avoid duplication

---

#  Disadvantages

1. Too many imports can confuse beginners
2. Name conflicts possible
3. Improper structure can make code complex

---

#  Summary

* Modules and packages help organize code
* Use `import`, `from`, and `as` for importing
* Packages allow hierarchical structure
* Improves readability and maintainability
