#  Abstraction 

##  Definition

**Abstraction** is an Object-Oriented Programming (OOP) concept that focuses on **hiding implementation details** and showing only the **essential features** of an object.

- In simple words:
It allows you to **use functionality without knowing how it works internally**.

---

#  Why Use Abstraction?

* Reduces complexity
* Improves code readability
* Enhances security
* Helps in large-scale application design

---

#  Real-Life Example

--> Driving a car:
* You use steering, brake, accelerator
* You don’t need to know how the engine works internally

---

#  How Abstraction is Achieved in Python

Python achieves abstraction using:

* **Abstract Classes**
* **Abstract Methods**
* `abc` module (Abstract Base Classes)

---

#  Abstract Class

* A class that **cannot be instantiated**
* Contains one or more abstract methods

```python id="g7p2l8"
from abc import ABC

class Shape(ABC):
    pass
```

---

#  Abstract Method

* A method declared but not implemented
* Must be implemented in child class

```python id="r4n1z6"
from abc import ABC, abstractmethod

class Shape(ABC):
    
    @abstractmethod
    def area(self):
        pass
```

---

#  Example

```python id="v8m5q3"
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

d = Dog()
d.sound()
```

---

#  Key Points

* Abstract class defines structure
* Child class must implement abstract methods
* Cannot create object of abstract class

---

#  Advantages

1. Hides unnecessary details
2. Reduces complexity
3. Improves maintainability
4. Increases code reusability

---

#  Disadvantages

1. Adds extra code
2. Can be complex for beginners
3. Requires proper design

---

#  Abstraction vs Encapsulation

| Feature | Abstraction         | Encapsulation     |
| ------- | ------------------- | ----------------- |
| Purpose | Hide implementation | Hide data         |
| Focus   | What to do          | How to protect    |
| Example | Abstract class      | Private variables |

---

#  Important Notes 

* Use `ABC` and `@abstractmethod`
* Abstract class cannot be instantiated
* Child class must override abstract methods
* Important pillar of OOP

---

#  Summary

* Abstraction hides implementation details
* Achieved using abstract classes and methods
* Improves code design and readability
* One of the four pillars of OOP


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def drive(self):
        print("Driving a vehicle")
    
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Starting the car engine")

class Bike(Vehicle):
    def start_engine(self):
        print("Starting the bike engine")

    
def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car = Car()
operate_vehicle(car)


bike = Bike()
operate_vehicle(bike)
````
