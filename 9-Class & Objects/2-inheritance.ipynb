#  Inheritance 

##  Definition

**Inheritance** is a feature of Object-Oriented Programming (OOP) that allows one class (child class) to acquire the properties and methods of another class (parent class).

- In simple words:
It allows **code reuse by inheriting features from another class**.

---

##  Basic Syntax

```python
class Parent:
    # parent class code

class Child(Parent):
    # child class code
```

---

##  Example

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()   # inherited method
d.bark()    # own method
```

### Output:

```
Animal speaks
Dog barks
```

---

#  Types of Inheritance

##  1. Single Inheritance

- One child inherits from one parent

```python
class A:
    pass

class B(A):
    pass
```

---

##  2. Multiple Inheritance

- One child inherits from multiple parents

```python
class A:
    pass

class B:
    pass

class C(A, B):
    pass
```

---

##  3. Multilevel Inheritance

- Chain of inheritance

```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass
```

---

##  4. Hierarchical Inheritance

- Multiple children from one parent

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass
```

---

##  5. Hybrid Inheritance

- Combination of multiple types

---

#  Method Overriding

- Child class can modify parent method

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")
```

---

#  Using `super()`

- Calls parent class methods

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)
```

---

#  Advantages

1. Code reuse
2. Reduces redundancy
3. Improves maintainability
4. Supports hierarchical classification

---

#  Disadvantages

1. Can make code complex
2. Tight coupling between classes
3. Debugging becomes harder

---

#  Important Notes 

* Child class inherits parent properties
* Use `super()` to access parent methods
* Supports multiple types of inheritance
* Enables method overriding
* Important concept in OOP

---

#  Summary

* Inheritance allows reuse of code
* Child class extends parent class
* Supports multiple types
* Helps build structured and scalable programs


````python
# parent class
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def display_info(self):
        print(f"Registration Year: {self.year}\nMade By: {self.make}\nModel: {self.model}")
````

````python
car1 = Car("Toyota", "Crown", 2020)
car1.display_info()
````

````python
# child class
class Tesla(Car):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity
    
    def display_battery_info(self):
        print(f"Battery Capacity: {self.battery_capacity} kWh")

tesla1 = Tesla("Tesla", "S", 2022, 100)
tesla1.display_info()
tesla1.display_battery_info()
````

````python
## Multiple Inheritance -> when a class inherits from more than one parent class
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"

class pet:
    def __init__(self, owner):
        self.owner = owner

class Dog(Animal, pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        pet.__init__(self, owner)
    
    def speak(self):
        return f"{self.name} says Woof!"

dog1 = Dog("Bob", "Alex")
print(dog1.speak())
print(f"Owner: {dog1.owner}")


````
