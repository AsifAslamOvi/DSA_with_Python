#  Object-Oriented Programming (OOP) 

##  Definition

**Object-Oriented Programming (OOP)** is a programming paradigm based on the concept of **objects**, which contain data (attributes) and functions (methods).

- In simple words:
It is a way of **organizing code using classes and objects**.

---

#  Basic Concepts of OOP

##  1. Class

A **class** is a blueprint or template for creating objects.

```python id="q9s8a1"
class Student:
    pass
```

---

##  2. Object

An **object** is an instance of a class.

```python id="x2l9b4"
s1 = Student()
```

---

##  3. Attributes and Methods

```python id="p8v3n2"
class Student:
    def __init__(self, name, age):
        self.name = name   # attribute
        self.age = age
    
    def display(self):     # method
        print(self.name, self.age)

s1 = Student("Asif", 21)
s1.display()
```


#  Types of Variables

##  Instance Variable

```python id="u5e2p9"
self.name = name
```

##  Class Variable

* Shared among all objects

##  Instance Method

##  Class Method


---

##  Static Method

```python id="d2x9r8"

```

---

#  Four Pillars of OOP

##  1. Encapsulation

* Wrapping data and methods together

```python id="z8q1w3"
class Person:
    def __init__(self):
        self.__age = 0   # private variable
```

---

##  2. Abstraction

* Hiding implementation details

```python id="g5y2n1"
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

---

##  3. Inheritance

* One class inherits another

```python id="b3t6k9"
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    pass
```

---

##  4. Polymorphism

* Same function, different behavior

```python id="n7f2x4"
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")
```

---

#  Method Overriding

```python id="v4m1c7"
class Parent:
    def show(self):
        print("Parent")

class Child(Parent):
    def show(self):
        print("Child")
```

---

#  Operator Overloading

```python id="r2q8s6"
class Point:
    def __init__(self, x):
        self.x = x
    
    def __add__(self, other):
        return self.x + other.x
```

````python
## instance variables and instance methods
class dog:
    ## constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

## create objects
dog1 = dog("jack", 2)
print(dog1)
print(dog1.name)
print(dog1.age)
````

````python
## define a class with instance variables and instance methods
class dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof!")

dog1 = dog("jon", 3)
print(dog1.name)
print(dog1.age)
dog1.bark()
````

````python
## Modeling a bank account
class BankAccount:
    def __init__(self, account_number, balance = 0):
        self.account_number = account_number
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount} on account {self.account_number}. New balance: {self.balance}")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdraw {amount} from account {self.account_number}. New balance: {self.balance}")
    
    def get_balance(self):
        return self.balance

account1 = BankAccount("1054900140001", 1000)
print(account1.account_number)
print(account1.balance)
account1.deposit(500)
account1.withdraw(200)
````

````python
print(account1.get_balance())
````
