#  Polymorphism 

##  Definition

**Polymorphism** is an important concept of Object-Oriented Programming (OOP) that allows the same function or method name to be used for different types of objects.

- In simple words:
It means **"one name, many forms"**.

---

#  Types of Polymorphism

##  1. Compile-Time Polymorphism

* Achieved using **method overloading**
* Same method name with different parameters

- Python does not support traditional method overloading directly

---

##  2. Runtime Polymorphism

* Achieved using **method overriding**
* Child class changes behavior of parent method

---

#  Function Polymorphism

- Same function works with different data types

```python id="nt5fx3"
print(len("Hello"))   # string
print(len([1, 2, 3])) # list
```

---

#  Operator Polymorphism

- Same operator behaves differently

```python id="2r0lkk"
print(5 + 3)        # addition
print("Hello " + "World")  # string concatenation
```

---

#  Method Overriding

```python id="czf5c7"
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
```

### Output:

```id="68qv9h"
Dog barks
```

---

#  Polymorphism with Multiple Classes

```python id="o2i0vf"
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

for animal in (Dog(), Cat()):
    animal.sound()
```

---

#  Duck Typing

- Python uses **duck typing**:

"If it looks like a duck and behaves like a duck, it is a duck."

```python id="1l0b3o"
class Bird:
    def fly(self):
        print("Flying")

class Airplane:
    def fly(self):
        print("Airplane flying")

for obj in (Bird(), Airplane()):
    obj.fly()
```

---

#  Advantages

1. Increases flexibility
2. Promotes code reuse
3. Makes code scalable
4. Simplifies interface design

---

#  Disadvantages

1. Can make code harder to understand
2. Debugging can be complex
3. Overuse reduces readability

---

#  Important Notes 

* Same method name, different behavior
* Achieved using overriding and duck typing
* Python supports runtime polymorphism
* Operator and function polymorphism also exist

---

#  Summary

* Polymorphism = one interface, many implementations
* Works using method overriding and duck typing
* Improves flexibility and reusability
* Core concept of OOP
