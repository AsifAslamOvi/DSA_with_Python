#  Polymorphism 

##  Definition

**Polymorphism** is an important concept of Object-Oriented Programming (OOP) that allows the same function or method name to be used for different types of objects.

- In simple words:
It means **"one name, many forms"**.

---

#  Types of Polymorphism

##  1. Compile-Time Polymorphism

* Achieved using **method overloading**
* Same method name with different parameters

- Python does not support traditional method overloading directly

---

##  2. Runtime Polymorphism

* Achieved using **method overriding**
* Child class changes behavior of parent method

---

#  Function Polymorphism

- Same function works with different data types

```python id="nt5fx3"
print(len("Hello"))   # string
print(len([1, 2, 3])) # list
```

---

#  Operator Polymorphism

- Same operator behaves differently

```python id="2r0lkk"
print(5 + 3)        # addition
print("Hello " + "World")  # string concatenation
```

---

#  Method Overriding

```python id="czf5c7"
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
```

### Output:

```id="68qv9h"
Dog barks
```

---

#  Polymorphism with Multiple Classes

```python id="o2i0vf"
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

for animal in (Dog(), Cat()):
    animal.sound()
```

---

#  Duck Typing

- Python uses **duck typing**:

"If it looks like a duck and behaves like a duck, it is a duck."

```python id="1l0b3o"
class Bird:
    def fly(self):
        print("Flying")

class Airplane:
    def fly(self):
        print("Airplane flying")

for obj in (Bird(), Airplane()):
    obj.fly()
```

---

#  Advantages

1. Increases flexibility
2. Promotes code reuse
3. Makes code scalable
4. Simplifies interface design

---

#  Disadvantages

1. Can make code harder to understand
2. Debugging can be complex
3. Overuse reduces readability

---

#  Important Notes 

* Same method name, different behavior
* Achieved using overriding and duck typing
* Python supports runtime polymorphism
* Operator and function polymorphism also exist

---

#  Summary

* Polymorphism = one interface, many implementations
* Works using method overriding and duck typing
* Improves flexibility and reusability
* Core concept of OOP


````python
'''
Method Overriding --> When a child class provides a specific implementation of a method that is already defined 
in its parent class, it is called method overriding. The child class's method will be called instead of the parent 
class's method when invoked on an instance of the child class.
'''

class Animal:
    def speak(self):
        return "Sound of the animal"

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

## Function that demonstrate polymorphism
def animal_speak(animal):
    print(animal.speak())

dog = Dog()
animal_speak(dog)

cat = Cat()
animal_speak(cat)
````

````python
## Polymorphism with functions and methods
class Shape:
    def area(self):
        return "The area of the shape"

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        import math
        return math.pi * self.radius ** 2

def print_area(Shape):
    print (f"The area is: {Shape.area()}")

rect = Rectangle(5, 3)
print_area(rect)

circle = Circle(4)
print_area(circle)

````

````python
## polymorphism with abstract base classes 
'''
Abstract Base Classes (ABCs) are a way to define interfaces in Python. They allow you to create classes that cannot 
be instantiated
'''
from abc import ABC, abstractmethod

## define an abstract base class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"

class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started"

def start_vehicle(Vehicle):
    print(Vehicle.start_engine())


car = Car()
start_vehicle(car)

motorcycle = Motorcycle()
start_vehicle(motorcycle)
````
