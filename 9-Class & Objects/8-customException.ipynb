#  Custom Exceptions 

##  Definition

A **custom exception** is a user-defined exception that is created by inheriting from Python’s built-in `Exception` class.

- In simple words:
It allows you to **define your own error types** for specific situations in your program.

---

#  Why Use Custom Exceptions?

* Handle application-specific errors
* Improve code readability
* Provide meaningful error messages
* Make debugging easier

---

#  Basic Syntax

```python id="q8n2z1"
class MyException(Exception):
    pass
```

- This creates a new exception class

---

#  Raising Custom Exception

```python id="m4t7c9"
class AgeError(Exception):
    pass

def check_age(age):
    if age < 18:
        raise AgeError("Age must be 18 or above")
    return "Access granted"

print(check_age(16))
```

---

#  Handling Custom Exception

```python id="v6p3k2"
class AgeError(Exception):
    pass

try:
    check_age(15)

except AgeError as e:
    print("Custom Error:", e)
```

---

#  Custom Exception with Constructor

```python id="r2s9h4"
class BalanceError(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

def withdraw(balance, amount):
    if amount > balance:
        raise BalanceError("Insufficient balance")
    return balance - amount
```

---

#  Example 

```python id="k5w8d1"
class NegativeNumberError(Exception):
    pass

try:
    num = int(input("Enter a number: "))
    
    if num < 0:
        raise NegativeNumberError("Negative number not allowed")
    
    print("You entered:", num)

except NegativeNumberError as e:
    print("Error:", e)
```

---

#  Built-in vs Custom Exceptions

| Feature | Built-in Exception    | Custom Exception           |
| ------- | --------------------- | -------------------------- |
| Source  | Python                | User-defined               |
| Example | ValueError, TypeError | AgeError, BalanceError     |
| Use     | General errors        | Specific application logic |

---

#  Advantages

1. Handles specific errors clearly
2. Improves readability
3. Easier debugging
4. Better program control

---

#  Disadvantages

1. Adds extra code
2. Requires proper design
3. Can be overused unnecessarily

---

#  Important Notes 

* Must inherit from `Exception` class
* Use `raise` to trigger exception
* Use `try-except` to handle it
* Useful in real-world applications
* Helps create meaningful error messages

---

#  Summary

* Custom exceptions are user-defined errors
* Created by inheriting `Exception`
* Raised using `raise`
* Handled using `try-except`
* Improves clarity and control in programs


````python
class Error(Exception):
    pass

class dobException(Error):
    pass
````

````python
year = int(input("Enter your year of birth: "))
age = 2026 - year

try:
    if age <= 32 and age >= 18:
        print("You are eligible to apply for the job.")
    else:
        raise dobException("You are not eligible to apply for the job.")

except dobException as e:
    print(e)
````
