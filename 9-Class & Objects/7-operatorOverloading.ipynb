#  Operator Overloading 

##  Definition

**Operator Overloading** is a feature in Python that allows operators (like `+`, `-`, `*`, etc.) to behave differently depending on the objects they are used with.

- In simple words:
It allows you to **define custom behavior for operators in user-defined classes**.

---

#  Why Use Operator Overloading?

* Makes code more readable
* Allows objects to behave like built-in types
* Provides flexibility in operations
* Useful in mathematical and logical operations

---

#  How It Works

Operator overloading is achieved using **magic methods (dunder methods)**.

| Operator | Magic Method  |
| -------- | ------------- |
| `+`      | `__add__`     |
| `-`      | `__sub__`     |
| `*`      | `__mul__`     |
| `/`      | `__truediv__` |
| `==`     | `__eq__`      |
| `<`      | `__lt__`      |
| `>`      | `__gt__`      |

---

#  Basic Example

```python id="q7m2c8"
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

n1 = Number(10)
n2 = Number(20)

result = n1 + n2
print(result.value)
```

### Output:

```id="p4s8d1"
30
```

#  Example with Multiple Operators

```python id="m5v9k3"
class Point:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Point(self.x + other.x)

    def __sub__(self, other):
        return Point(self.x - other.x)

p1 = Point(10)
p2 = Point(5)

print((p1 + p2).x)
print((p1 - p2).x)
```

#  Example with String Representation

```python id="y2f8w6"
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student: {self.name}"
```

````python
class vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, other):
        return vector(self.x * other.x, self.y * other.y)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
    def __repr__(self):
        return f"vector({self.x}, {self.y})"

v1 = vector(1, 2)
v2 = vector(3, 4)

print(v1 + v2)
print(v1 - v2)
print(v1 * v2)
print(v1 == v2)
print(v1)
print(repr(v1))
````

````python
## Overloading Operators for complex number
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img
    
    def __add__(self, other):
        return Complex(self.real + other.real, self.img + other.img)
    
    def __sub__(self, other):
        return Complex(self.real - other.real, self.img - other.img)

    def __mul__(self, other):
        return Complex(self.real * other.real - self.img * other.img, self.real * other.img + self.img * other.real)
    
    def __truediv__(self, other):
        denominator = other.real ** 2 + other.img ** 2
        real_part = (self.real * other.real + self.img * other.img) / denominator
        img_part = (self.img * other.real - self.real * other.img) / denominator
        return Complex(real_part, img_part)
    
    def __repr__(self):
        return f"Complex({self.real}, {self.img})"
    
c1 = Complex(1, 2)
c2 = Complex(3, 4)
print(c1 + c2)
print(c1 - c2)
print(c1 * c2)
print(c1 / c2)
print(c1)
````
