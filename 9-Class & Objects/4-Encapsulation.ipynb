#  Encapsulation 

##  Definition

**Encapsulation** is an Object-Oriented Programming (OOP) concept that involves **wrapping data (variables) and methods (functions) into a single unit (class)** and restricting direct access to some components.

- In simple words:
It is used to **hide internal data and control access to it**.

---

#  Why Use Encapsulation?

* Protects data from unauthorized access
* Improves code security
* Helps maintain control over data
* Makes code modular and maintainable

---

#  Basic Example

```python id="k9x2fm"
class Student:
    def __init__(self, name, marks):
        self.name = name        # public variable
        self.__marks = marks    # private variable

    def display(self):
        print(self.name, self.__marks)

s = Student("Asif", 90)
s.display()
```

---

#  Access Modifiers in Python

##  1. Public Members

* Accessible from anywhere

```python id="g3s8k1"
self.name = "Asif"
```

---

##  2. Protected Members

* Indicated with single underscore `_`
* Should not be accessed directly (convention)

```python id="d2n7q4"
self._age = 21
```

---

##  3. Private Members

* Indicated with double underscore `__`
* Cannot be accessed directly outside class

```python id="q8p4v1"
self.__marks = 90
```

---

#  Accessing Private Variables

##  Direct Access (Error)

```python id="n2z6k3"
print(s.__marks)  # AttributeError
```

---

##  Access via Method

```python id="h6r3x9"
def get_marks(self):
    return self.__marks
```

---

##  Getter and Setter Methods

###  Getter

```python id="t1f7p2"
def get_marks(self):
    return self.__marks
```

###  Setter

```python id="y5m8d4"
def set_marks(self, value):
    self.__marks = value
```

---

##  Example with Getter & Setter

```python id="p3s9c7"
class Student:
    def __init__(self):
        self.__marks = 0

    def set_marks(self, marks):
        if marks >= 0:
            self.__marks = marks

    def get_marks(self):
        return self.__marks

s = Student()
s.set_marks(85)
print(s.get_marks())
```

#  Name Mangling 

- Python changes private variable name internally:

```id="a8r2k6"
__marks → _ClassName__marks
```

### Example:

```python id="f4j1l0"
print(s._Student__marks)
```

````python
## protected variable
'''
protected variable --> A protected variable is a variable that is intended to be accessed only within the class 
and its subclasses. It is denoted by a single underscore prefix (e.g., _variable). While it can be accessed from 
outside the class, it is generally discouraged to do so, as it indicates that the variable is meant for internal 
use only.
'''
class Person:
    def __init__(self, name, age, gender):
        self.name = name # public variable
        self._age = age # protected variable
        self._gender = gender # protected variable

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)
'''
super() --> The super() function is used to call a method from the parent class. In this case, it is used to call 
the __init__ method of the parent class. This allows the Employee class to inherit the properties of the Person class,
including the protected variables
'''

employee = Employee("Asif", 23, "Male")
print(employee.name) # Accessing public variable
print(employee._age) # Accessing protected variable from subclass
print(employee._gender) # Accessing protected variable from subclass
````

````python
### Encapsulation with getter and setter methods
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age

    # Getter method for age
    def get_age(self):
        return self.__age

    # Setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid age. Age must be a positive integer.")

person = Person("Asif", 23)
print(person.name) # Accessing public variable

print(person.get_age()) # Accessing private variable through getter method

person.set_age(25) # Modifying private variable through setter method
print(person.get_age()) # Accessing modified private variable through getter method

person.set_age(-5) # Attempting to set invalid age
````
