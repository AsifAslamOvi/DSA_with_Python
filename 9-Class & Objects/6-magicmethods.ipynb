#  Magic Methods (Dunder Methods) 

##  Definition

**Magic methods** (also called **dunder methods**, short for “double underscore”) are special methods in Python that start and end with double underscores (`__`).

- In simple words:
They allow you to **define how objects behave with built-in operations** like addition, printing, comparison, etc.

---

#  Why Use Magic Methods?

* Customize object behavior
* Enable operator overloading
* Integrate with Python built-in functions
* Make objects behave like built-in types

---

#  Common Magic Methods

##  1. `__init__` (Constructor)

* Called when object is created

```python id="j2m9q1"
class Student:
    def __init__(self, name):
        self.name = name
```

---

##  2. `__str__` (Readable String)

* Defines output for `print()`

```python id="q8r1x3"
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student: {self.name}"

s = Student("Asif")
print(s)
```

---

##  3. `__repr__` (Official Representation)

* Used for debugging

```python id="t5v2k7"
def __repr__(self):
    return f"Student('{self.name}')"
```

---

##  4. `__add__` (Addition Operator)

```python id="b7n4d9"
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value
```

---

##  5. `__len__` (Length)

```python id="z3p6c2"
class MyList:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)
```

---

##  6. `__eq__` (Equality Check)

```python id="k1r9f5"
def __eq__(self, other):
    return self.value == other.value
```

---

##  7. `__lt__` and `__gt__` (Comparison)

```python id="h6y2v8"
def __lt__(self, other):
    return self.value < other.value

def __gt__(self, other):
    return self.value > other.value
```

---

##  8. `__getitem__` (Indexing)

```python id="m9c4w1"
def __getitem__(self, index):
    return self.items[index]
```

---

##  9. `__setitem__` (Assignment)

```python id="x5d8p7"
def __setitem__(self, index, value):
    self.items[index] = value
```

---

##  10. `__del__` (Destructor)

```python id="n2a7q3"
def __del__(self):
    print("Object deleted")
```

---

#  Operator Overloading Example

```python id="u8s3f6"
class Point:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Point(self.x + other.x)

p1 = Point(3)
p2 = Point(4)

p3 = p1 + p2
print(p3.x)
```

---

#  Advantages

1. Makes objects behave like built-in types
2. Improves readability
3. Enables operator overloading
4. Provides powerful customization

---

#  Disadvantages

1. Can make code confusing if overused
2. Harder to debug
3. Requires good understanding

---

#  Important Notes 

* Always use double underscores (`__method__`)
* Used for operator overloading and customization
* `__init__` is the most commonly used
* `__str__` improves print output
* Important for advanced Python programming

---

#  Summary

* Magic methods define special behavior of objects
* Enable operator overloading and customization
* Start and end with double underscores
* Used in many built-in Python operations


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("Asif", 23)
print(person)
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"
    
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person = Person("Asif", 23)
print(person)
print(repr(person))
````
