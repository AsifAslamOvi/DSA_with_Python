#  `map()` Function in Python

##  Definition

The `map()` function in Python is used to apply a given function to each item of an iterable (such as a list, tuple, etc.) and return a map object (iterator).

 In simple words:
It **transforms all elements of a collection using a function**.

---

##  Syntax

```python
map(function, iterable)
```

* `function` → The function to apply to each element
* `iterable` → A sequence (list, tuple, etc.)

---

##  Return Value

* Returns a **map object (iterator)**
* You usually convert it to:

  * `list()`
  * `tuple()`
  * `set()`

---

##  Basic Example

```python
def square(x):
    return x ** 2

nums = [1, 2, 3, 4]
result = map(square, nums)

print(list(result))
```

### Output:

```
[1, 4, 9, 16]
```

---

##  Using `map()` with Lambda

```python
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))

print(result)
```

### Output:

```
[2, 4, 6, 8]
```

 Very common usage in exams and interviews

---

##  Using Multiple Iterables

```python
a = [1, 2, 3]
b = [4, 5, 6]

result = list(map(lambda x, y: x + y, a, b))
print(result)
```

### Output:

```
[5, 7, 9]
```

 Works element-wise:

* (1+4), (2+5), (3+6)

---

##  Convert Data Types Using `map()`

```python
nums = ["1", "2", "3"]
result = list(map(int, nums))

print(result)
```

### Output:

```
[1, 2, 3]
```

---

##  Using `map()` with Strings

```python
names = ["asif", "ovi", "python"]
result = list(map(str.upper, names))

print(result)
```

### Output:

```
['ASIF', 'OVI', 'PYTHON']
```

---

##  `map()` vs Loop

### Using Loop:

```python
nums = [1, 2, 3]
result = []

for x in nums:
    result.append(x * 2)
```

### Using `map()`:

```python
result = list(map(lambda x: x * 2, nums))
```

 `map()` is shorter and cleaner

---

##  Advantages

1) Code is shorter and cleaner
2) Faster than loops (in many cases)
3) Works well with lambda functions
4) Supports multiple iterables

---

##  Disadvantages

1) Less readable for beginners
2) Only works for simple transformations
3) Returns iterator (must convert to list for reuse)

---

##  Important Notes 

* `map()` does NOT modify original data
* It returns a **new transformed iterator**
* Stops when the shortest iterable ends
* Often used with:

  * `lambda`
  * `list()`
  * `filter()`

---

##  Real-Life Example

```python
prices = [100, 200, 300]

# Add 10% tax
final_prices = list(map(lambda x: x * 1.1, prices))

print(final_prices)
```

### Output:

```
[110.0, 220.0, 330.0]
```

---

##  Nested Use (Advanced)

```python
nums = [1, 2, 3]
result = list(map(lambda x: x**2 + 1, nums))

print(result)
```

---

#  Summary

* `map()` applies a function to every element of an iterable
* Returns a map object (iterator)
* Often used with lambda for short operations
* Can work with multiple iterables
* More concise than loops for transformations


````python
def square(x):
    return x * x

square(10)
````

````python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

squared_numbers = list(map(square, numbers))
print(squared_numbers)
````

````python
## lambda function with map 
squared_numbers_lambda = list(map(lambda x: x**2 , numbers))
print(squared_numbers_lambda)
````

````python
### multiple iterable with map
numbers1 = [1, 2, 3, 4]
numbers2 = [5, 6, 7, 8]
summed_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))
print(summed_numbers)
````

````python
## map() to convert a list of strings to integer 
string_numbers = ['1', '2', '3', '4']
int_numbers = list(map(int, string_numbers))
print(int_numbers)
````

````python
words = ["apple", "banana", "cherry", "date", "elderberry"]
upper_words = list(map(str.upper, words))
print(upper_words)
````

````python
def get_name(person):
    return person['name']

people = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35}
]

names = list(map(get_name, people))
print(names)
````
