# Lambda Functions in Python

## Definition

A **lambda function** is a small anonymous (unnamed) function in Python that is defined using the `lambda` keyword.

--> It can take any number of arguments but can have **only one expression**.

---

## Syntax

```python
lambda arguments: expression
```

* `lambda` → keyword to define the function
* `arguments` → input parameters
* `expression` → single expression that is evaluated and returned

---

## Basic Example

```python
add = lambda a, b: a + b
print(add(3, 5))
```

### Output:

```
8
```

## Equivalent to:

```python
def add(a, b):
    return a + b
```

---

## Key Characteristics

### 1. Anonymous Function

* No function name required

### 2. Single Expression Only

* Cannot use multiple statements
* No loops, assignments, or multiple lines

## Invalid:

```python
lambda x: x = x + 1
```

---

### 3. Automatically Returns Value

* No need to write `return`

---

### 4. Multiple Arguments Allowed

```python
multiply = lambda x, y, z: x * y * z
print(multiply(2, 3, 4))  # 24
```

---

## Why Use Lambda?

==> Used for:

* Short, simple operations
* Temporary (one-time) functions
* Functional programming (`map`, `filter`, `sorted`)

---

# Common Use Cases

## 1. Using `map()`

```python
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
print(squared)
```

### Output:

```
[1, 4, 9, 16]
```

---

## 2. Using `filter()`

```python
nums = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)
```

### Output:

```
[2, 4]
```

---

## 3. Using `sorted()`

```python
students = [("Alice", 25), ("Bob", 20), ("Charlie", 23)]

sorted_list = sorted(students, key=lambda x: x[1])
print(sorted_list)
```

### Output:

```
[('Bob', 20), ('Charlie', 23), ('Alice', 25)]
```

---

## 4. Passing Lambda as Argument

```python
def apply_func(x, func):
    return func(x)

print(apply_func(5, lambda x: x * 2))
```

---

## Lambda vs Normal Function

| Feature    | Lambda      | Normal Function     |
| ---------- | ----------- | ------------------- |
| Name       | Anonymous   | Named               |
| Syntax     | Short       | Longer              |
| Complexity | Simple only | Complex logic       |
| Return     | Implicit    | Explicit (`return`) |

---

## Advantages

 1) Short and concise
 2) No need to define full function
 3) Useful for one-time use
 4) Works well with functional tools

---

## Disadvantages

1)  Hard to read for complex logic
2)  Limited to one expression
3)  Cannot include statements like loops, assignments

---

## Important Notes 

* Lambda functions are **not always better** than normal functions
* Use lambda only when:

  * Function is simple
  * Used once
* Otherwise, use `def` for readability

---

## Real-Life Example

```python
data = [("A", 3), ("B", 1), ("C", 2)]

# Sort by second value
print(sorted(data, key=lambda x: x[1]))
```

---

## Nested Lambda

```python
add = lambda x: (lambda y: x + y)
print(add(5)(3))  # 8
```

---

## Lambda with Conditional

```python
check = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check(4))  # Even
```

---

# Summary

* Lambda = small, anonymous function
* Defined using `lambda` keyword
* Can take multiple inputs but only one expression
* Commonly used with `map()`, `filter()`, `sorted()`
* Best for short, simple operations


````python
## Syntax 
lambda arguments : expression
````

````python
def add(a, b):
    return a + b
````

````python
add(2, 3)
````

````python
addition = lambda a, b : a + b
print(type(addition))
print(addition(5, 7))
````

````python
def even(num):
    if num % 2 == 0:
        return True
    return False

print(even(4))
print(even(5))
````

````python
even1 = lambda num : num % 2 == 0
print(even1(4))
print(even1(5))
````

````python
# map -> Applies a function to all items in an iterable (like a list) and returns an iterator of the results.
numbers = [1, 2, 3, 4, 5, 6]
def square(numbers):
    return numbers ** 2

squared_numbers = map(square, numbers)
print(list(squared_numbers))
````

````python
squared_numbers_lambda = map(lambda x : x ** 2, numbers)
print(list(squared_numbers_lambda))
````
