#  File Operations 

##  Definition

File operations in Python refer to **creating, reading, writing, and modifying files** using built-in functions.

- Files help store data permanently on disk.

---

#  Types of File Operations

##  1. Opening a File

Python uses the `open()` function.

```python
file = open("example.txt", "mode")
```

###  File Modes:

| Mode | Meaning           |
| ---- | ----------------- |
| `r`  | Read (default)    |
| `w`  | Write (overwrite) |
| `a`  | Append            |
| `x`  | Create new file   |
| `rb` | Read binary       |
| `wb` | Write binary      |

---

##  Example:

```python
file = open("data.txt", "r")
```

---

#  2. Reading a File

##  Read Entire File

```python
file = open("data.txt", "r")
content = file.read()
print(content)
file.close()
```

---

##  Read Line by Line

```python
file = open("data.txt", "r")

for line in file:
    print(line)

file.close()
```

---

##  Read One Line

```python
file = open("data.txt", "r")
print(file.readline())
file.close()
```

---

##  Read All Lines as List

```python
file = open("data.txt", "r")
lines = file.readlines()
print(lines)
file.close()
```

---

#  3. Writing to a File

##  Write Mode (`w`)

 Overwrites existing content

```python
file = open("data.txt", "w")
file.write("Hello World")
file.close()
```

---

##  4. Appending to a File

##  Append Mode (`a`)

- Adds content without deleting existing data

```python
file = open("data.txt", "a")
file.write("\nNew Line Added")
file.close()
```

---

#  5. Creating a New File

##  Mode `x`

- Creates file, gives error if file exists

```python
file = open("newfile.txt", "x")
file.write("This is a new file")
file.close()
```

---

#  6. Closing a File

```python
file.close()
```

- Important to free system resources

---

#  Best Practice (Using `with`)

- Automatically closes file

##  Reading

```python
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```

---

##  Writing

```python
with open("data.txt", "w") as file:
    file.write("Hello Python")
```

---

#  File Methods

| Method        | Description               |
| ------------- | ------------------------- |
| `read()`      | Reads full file           |
| `readline()`  | Reads one line            |
| `readlines()` | Reads all lines into list |
| `write()`     | Writes text               |
| `close()`     | Closes file               |

---

#  File Pointer (Important Concept)

* Pointer shows current position in file
* Moves as you read/write

```python
file = open("data.txt", "r")
print(file.tell())  # current position
file.seek(0)        # move pointer to start
file.close()
```

---

#  Binary File Operations

```python
file = open("image.jpg", "rb")
data = file.read()
file.close()
```

---

#  Advantages

1. Stores data permanently
2. Easy data handling
3. Useful for large datasets
4. Supports multiple formats

---

#  Disadvantages

1. Slower than memory operations
2. Risk of data corruption
3. Requires proper handling (close file)

---

#  Important Notes 

* Always close file after use
* Use `with open()` for safety
* `w` mode overwrites data
* `a` mode preserves old data
* Files are stored in disk permanently

---

#  Summary

* File operations include read, write, append, create
* Use `open()` with modes (`r`, `w`, `a`, `x`)
* `with open()` is best practice
* Files help store permanent data


````python
## Read a whole file
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
````

````python
## Read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) # strip() removes leading/trailing whitespace
    
````

````python
## writing to a file(overwrite)

with open('example.txt', 'w') as file:
    file.write("This is a new line.\n")
    file.write("This will overwrite the existing content.\n")

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
    
````

````python
## write to a file without overwriting (append mode)
with open('example.txt', 'a') as file:
    file.write("This line will be appended to the existing content.\n")
    
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
````

````python
## writing a list of lines to a file
lines = ["First line\n", "Second line\n", "Third line\n"]

with open('example.txt', 'a') as file:
    file.writelines(lines) # writelines() writes a list of lines to the file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
````

````python
## Binary file operations
data = b'\x00\x01\x02\x03\x04' # some binary data
with open('example.bin', 'wb') as file:
    file.write(data) # write binary data to the file

with open('example.bin', 'rb') as file:
    binary_content = file.read()
    print(binary_content)
````

````python
### Read the content from a source text file and write it to a new file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('copy_of_example.txt', 'w') as dest_file:
    dest_file.write(content)
````

````python
## Read a text file and count the number of lines, words, and characters
with open('example.txt', 'r') as file:
    content = file.read()
    lines = content.splitlines() # splitlines() splits the content into lines
    words = content.split() # split() splits the content into words based on whitespace
    num_lines = len(lines)
    num_words = len(words)
    num_characters = len(content)

print(f"Number of lines: {num_lines}")
print(f"Number of words: {num_words}")
print(f"Number of characters: {num_characters}")
````

````python
## Read a text file and count the number of lines, words, and characters using function
def count_file_content(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        lines = content.splitlines()
        words = content.split()
        num_lines = len(lines)
        num_words = len(words)
        num_characters = len(content)
    return num_lines, num_words, num_characters


file_path = 'example.txt'
lines, words, characters = count_file_content(file_path)
print(f"Number of lines: {lines}")
print(f"Number of words: {words}")
print(f"Number of characters: {characters}")
````

````python
## writing and reading a file
with open('example.txt', 'w+') as file:
    # w+ -> open for reading and writing, truncating the file first
    file.write("This is a line in the file.\n")
    file.write("This file is used for demonstrating read and write operations.\n")
    file.seek(0) # Move the file pointer back to the beginning of the file
    content = file.read() # Read the content of the file
    print(content)
````
