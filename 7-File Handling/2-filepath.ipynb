#  File Path 

##  Definition

A **file path** is the location of a file or directory in a computer’s file system.

- In simple words:
It tells Python **where the file is stored**.

---

##  Types of File Paths

##  1. Absolute Path

* Full path from the root directory
* Starts from the main system directory

### Example:

```python
"C:/Users/Asif/Documents/data.txt"
```

- Always points to the same location

---

##  2. Relative Path

* Path relative to the current working directory

### Example:

```python
"data.txt"
"folder/data.txt"
```

- Depends on where your program is running

---

#  Working with File Paths

##  Using `os` Module

```python
import os

print(os.getcwd())  # current working directory
```

---

##  Joining Paths

```python
import os

path = os.path.join("folder", "file.txt")
print(path)
```

- Handles OS differences automatically

---

##  Checking File Existence

```python
import os

print(os.path.exists("data.txt"))
```

---

##  Get File Info

```python
import os

print(os.path.isfile("data.txt"))
print(os.path.isdir("folder"))
```

---

#  Using `pathlib` Module (Modern Way)

```python
from pathlib import Path

path = Path("data.txt")

print(path.exists())
print(path.is_file())
```

---

##  Create Path

```python
path = Path("folder") / "file.txt"
print(path)
```

---

#  File Path Examples

##  Windows Path

```python
"C:\\Users\\Asif\\file.txt"
```

- Double `\\` is required

---

##  Raw String (Recommended)

```python
r"C:\Users\Asif\file.txt"
```

- Avoids escape issues

---

##  Linux/Mac Path

```python
"/home/asif/file.txt"
```

---

#  Common File Path Operations

| Operation         | Function           |
| ----------------- | ------------------ |
| Current directory | `os.getcwd()`      |
| Change directory  | `os.chdir()`       |
| Check existence   | `os.path.exists()` |
| Join path         | `os.path.join()`   |

---

#  Example: Open File with Path

```python
file_path = "data.txt"

with open(file_path, "r") as file:
    content = file.read()
    print(content)
```

---

#  Advantages

1. Helps locate files correctly
2. Supports cross-platform development
3. Works with both files and directories

---

#  Disadvantages

1. Wrong path causes errors
2. OS differences can create issues
3. Relative paths may break if directory changes

---

#  Important Notes

* Absolute path = full location
* Relative path = depends on current directory
* Use `os.path.join()` for portability
* Prefer `pathlib` for modern Python
* Use raw strings for Windows paths

---

#  Summary

* File path defines location of a file
* Two types: absolute and relative
* Use `os` or `pathlib` to handle paths
* Important for file operations


````python
import os 
cwd = os.getcwd()
print(f"Current working directory: {cwd}")
````

````python
## create a new directory called 
new_dir = "package"
os.mkdir(new_dir)
print(f"Directory '{new_dir}' created successfully.")
````

````python
## listing files and directories 
items = os.listdir(cwd)
print("Items in the current directory:")
for item in items:
    print(item)
````

````python
## joining paths
dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(dir_name, file_name)
print(f"Full path: {full_path}")
````

````python
dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(os.getcwd(),dir_name, file_name)
print(f"Full path: {full_path}")
````

````python
path = 'example2.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' does not exist")
````

````python
## checking if a path is a file or directory
import os
path = 'example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' does not exist.")
````

````python
## getting the absolute path of a file
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print(f"Absolute path: {absolute_path}")
````
